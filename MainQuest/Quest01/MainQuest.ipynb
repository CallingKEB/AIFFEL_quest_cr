```python
import random           # random 모듈

class Account:

    account_count = 0       # 계좌 갯수 변수 생성 및 초기화

    # 생성자 메서드
    def __init__(self, name, balance):
        self.bank = 'SC은행'            # 은행이름: SC은행
        self.name = name                # 예금주
        self.balance = balance          # 초기 잔액
        self.account_num = self.create_account_num()        # 계좌번호 생성
        self.deposit_count = 0          # 입금 횟수
        self.deposit_history_list = []  # 입금 내역 저장
        self.withdraw_history_list = [] # 출금 내역 저장

        Account.account_count += 1      # 계좌 객체가 생성되면 계좌 개수 증가

    # 계좌번호 생성 메서드
    def create_account_num(self):
        num1 = random.randint(100,999)          # 3자리
        num2 = random.randint(10, 99)           # 2자리
        num3 = random.randint(100000, 999999)   # 6자리
        return f"{num1}-{num2}-{num3}"             # 계좌번호 3자리-2자리-6자리 랜덤 (111-11-111111)
        #return str(num1) + '-' + str(num2) + '-' + str(num3)             # 계좌번호 3자리-2자리-6자리 랜덤 (111-11-111111)

    # 클래스 메서드 - 생성된 계좌 객체 개수
    @classmethod
    def get_account_num(cls):                   # cls: class 메서드 사용 (@classmethod)
        print(cls.account_count)

    # 입금 메서드 - 1원 이상만 입금 가능
    def deposit(self, cash):
        if cash >= 1:
            self.balance += cash
            self.deposit_count += 1
            self.deposit_history_list.append((self.deposit_count, '입금', cash, self.balance))    # 입금 내역 저장

            if self.deposit_count % 5 == 0:         # 입금 횟수가 5회일 때 이자 지급(6회, 12회 ...)
                interest = self.balance * 0.01      # 잔고의 1%
                self.balance += int(interest)       # 소수점 버림
                self.deposit_count += 1
                self.deposit_history_list.append((self.deposit_count, '이자지급', int(interest), self.balance))    # 이자 내역 저장
                print(f"{int(interest)}의 이자가 입금되었습니다.")

            else:
                if self.deposit_count % 6 == 0:
                    self.deposit_count = (self.deposit_count % 5 ) * 5 + (self.deposit_count % 5)   # 6회 이후 부분이 해결되지 않았음 (12회에 이자지급이 되어야 하는데 11회에 이자지급이 되고 있음)

        else:
            print("입금은 최소 1원 이상만 가능합니다.")

   # 출금 메서드 - 계좌 잔고만큼만 출금 가능
    def withdraw(self, cash):
        if self.balance >= cash:
            self.balance -= cash
            self.withdraw_history_list.append((len(self.withdraw_history_list) + 1, cash, self.balance))   # 출금 내역 저장

        else:
            print("계좌 잔고 이상으로 출금할 수 없습니다.")

    # 정보 출력 메서드
    def display_info(self):
        balance_str = '{:,}'.format(self.balance)       # 잔고에 세 자리마다 쉼표 추가
        print(f"은행이름: {self.bank}, 예금주: {self.name}, 계좌번호: {self.account_num}, 잔고: {balance_str}") # 잔액을 세자리마다 쉼표 추가

    # 입금 내역 출력 메서드
    def deposit_history(self):
        for count, transit_type, cash, balance in self.deposit_history_list:
            print(f"{count}회: {transit_type}       금액: {cash}    잔액: {balance}")


            #if transit_type == "입금":
            #    print(f"{count}회: 입금          금액: {cash}    잔액: {balance}")
            #elif transit_type == "이자지급":
            #    print(f"{count}회: 이자지급      금액: {cash}      잔액: {balance}")


            # 오류! >> print(f"{cash[0]}회: 출금  금액: {cash[1]}원  잔액: {cash[2]}원")
            # 횟수 및 입금(이자지급) 오류! >> print(f"{count}회: 입금    금액: {cash}    잔액: {balance}")

    # 출금 내역 출력 메서드
    def withdraw_history(self):
        for count, cash, balance in self.withdraw_history_list:
            print(f"{self.deposit_count + count}회: 출금           금액: {cash}    잔액: {balance}")
        #for count, cash, balance in self.withdraw_history_list:
            #print(f"{count}회: 출금           금액: {cash}    잔액: {balance}")

# 여러 계좌 생성 & 리스트 저장
first_account = Account("차정은", 1000)
second_account = Account("박광석", 30000000)
third_account = Account("조웅제", 50000000)

accounts = [first_account, second_account, third_account]

# 객체 순회 - 잔고 100만원 이상인 고객 정보만 출력
print("생성된 계좌의 총 개수: ", len(accounts))
for acc in accounts:
    if acc.balance >= 1000000:
        acc.display_info()

print()

# 입출금 내역 출력
my_account = Account("차정은", 1000)

for i in range(10):
    my_account.deposit(1000)

my_account.deposit(-1000)
my_account.withdraw(3000000)
my_account.withdraw(3000)

print()

my_account.deposit_history()
print('-'*50)
my_account.withdraw_history()




```
