# 프로젝트 1 (사각형 넓이 구하기)

```python
# 사각형 넓이를 구하는 클래스!
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')

        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    # 직사각형
    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    # 평행사변형
    def par(self):
        base, height = map(int, input('밑변, 높이를 입력하세요. 예시: 밑변,높이\n>>>').split(','))
        area = base * height
        result = '평행사변형의 넓이는: ' + str(area)
        return result

    # 사다리꼴
    def trape(self):
        base_top, base_bottom, height = map(int, input('윗변, 아랫변, 높이를 입력하세요. 예시: 윗변,아랫변,높이\n>>>').split(','))
        area = ((base_top + base_bottom) * height) / 2
        result = '사다리꼴의 넓이는: ' + str(area)
        return result

a = Square()  # 객체 생성
a.trape()       # 사다리꼴의 넓이 3,2,1 입력시 넓이 2.5가 나와야 한다
```

# 프로젝트 2 (키오스크 만들기)

```python
# 전체 메뉴와 가격
menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]

class Kiosk:
    def __init__(self):
        self.menu = menu
        self.price = price
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트
        self.price_sum = 0

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        n = 0
        while n < 1 or len(menu) < n:
            n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력
            # 메뉴판에 있는 음료 번호일 때
            if 1 <= n & n <= len(menu):
                self.price_sum = self.price[n-1]  # 선택 음료의 가격
                print(self.menu[n-1], ' : ', self.price[n-1], '원')  # 메뉴
            # 메뉴판에 없는 번호일 때
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 음료 온도 물어보기
        t = 0  # 기본값을 넣어주고
        while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
            t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
            if t == 1:
                self.temp = 'Hot'
                break
            elif t == 2:
                self.temp = 'ICE'
                break
            else:
                print("1과 2 중 하나를 입력하세요.\n")
        print('주문 음료', self.temp, self.menu[n-1], ' : ', self.price[n-1], '원')  # 온도 속성을 추가한 주문 결과 출력

        # 첫 주문일 때 처리!
        # self.order_menu.append(self.menu[n-1])   # 가격 리스트에 추가 (이렇게 하면 메뉴만 찍혔다!)
        self.order_menu.append(f'{self.temp} {self.menu[n-1]}')   # 가격 리스트에 추가 + 온도도 함께 추가
        self.order_price.append(self.price[n-1])  # 주문 리스트에 추가
        # self.price_sum += self.price[n-1]       # 첫 주문일 때는 합계 금액을 업데이트하니 계속 맞지 않는 금액이 결과로 도출되었다!

        # 추가 주문 또는 지불
        while n != 0:  # 지불을 선택하기 전까지 반복합니다.
            print()  # 줄 바꾸면서
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))  # 추가 주문 또는 지불
            if n > 0 and n < len(self.menu) + 1:
                # 추가 음료의 온도를 입력받고, 가격 리스트, 주문 리스트, 합계 금액을 업데이트
                while True:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = 'Hot'
                        break
                    elif t == 2:
                        self.temp = 'ICE'
                        break
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")

                #추가 주문건 처리!
                #self.order_menu.append(self.menu[n-1])   # 가격 리스트에 추가
                self.order_menu.append(f'{self.temp} {self.menu[n-1]}')   # 가격 리스트에 추가 + 온도도 함께 추가
                self.order_price.append(self.price[n-1])  # 주문 리스트에 추가
                self.price_sum += self.price[n-1]         # 합계 금액 업데이트

                print('추가 주문 음료', self.temp, self.menu[n-1], ':', self.price[n-1], '원\n', '합계 : ', self.price_sum, '원')
            else :
                if n == 0 :  # 지불을 입력하면
                    print("주문이 완료되었습니다.")
                    print(self.order_menu, self.order_price)  # 확인을 위한 리스트를 출력합니다.
                else :  # 없는 번호를 입력할 때
                    print("없는 메뉴입니다. 다시 주문해 주세요.")

    # 지불
    def pay(self):
        while True:
            b = int(input("지불수단을 고르세요 (현금:1, 카드:2): "))
            if b == 1:
                print("직원을 호출하겠습니다.")
                break
            elif b == 2:
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                break
            else:
                print("1과 2 중 하나를 입력하세요.\n")

    # 주문서 출력
    def table(self):
        def border_top():   # 외곽
            print('⟝' + '-' * 30 + '⟞')
            for i in range(5):
                print('|' + ' ' * 31 + '|')

        border_top()

        # 주문 상품명 : 해당 금액
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i],'원')

        print('합계 금액 :', self.price_sum,'원')

        def border_under():   # 외곽
            for i in range(5):
                print('|' + ' ' * 31 + '|')
            print('⟝' + '-' * 30 + '⟞')

        border_under()

### 출력 ###
a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```

# 프로젝트 2 (키오스크 만들기 - 확장버전)

```python
'''
확장 해보기!
1. 메뉴와 가격 리스트를 다차원 리스트로 변경하기
2. 메뉴출력 enumerate() 함수 활용하기
3. datetime 모듈로 주문표에 주문 일시 추가하기
4. 데코레이터로 주문표 양식 출력하기
'''

from datetime import datetime

# 전체 메뉴와 가격(확장 1. 다차원 리스트 사용)
menu = ['americano', 2000],['latte', 3000], ['mocha', 3000], ['yuza_tea', 2500], ['green_tea', 2500], ['choco_latte', 3000]

class Kiosk:
    def __init__(self):
        self.menu = menu
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트
        self.price_sum = 0
        self.order_time = None  # 주문 시각 추가

    # 메뉴 출력 메서드(확장 2. enumerate 사용)
    def menu_print(self):
        for i, (item, price) in enumerate(self.menu, 1):
            print(i, item, ':', price,'원')

    # 주문 메서드
    def menu_select(self):
        n = 0
        while n < 1 or len(menu) < n:
            n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력
            # 메뉴판에 있는 음료 번호일 때
            if 1 <= n & n <= len(menu):
                selected_menu = self.menu[n-1]   # 선택된 음료
                print(selected_menu[0], ' : ', selected_menu[1], '원')  # 메뉴
            # 메뉴판에 없는 번호일 때
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 음료 온도 물어보기
        t = 0  # 기본값을 넣어주고
        while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
            t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
            if t == 1:
                self.temp = 'Hot'
                break
            elif t == 2:
                self.temp = 'ICE'
                break
            else:
                print("1과 2 중 하나를 입력하세요.\n")
        #print('주문 음료', self.temp, self.menu[n-1], ' : ', self.price[n-1], '원')  # 온도 속성을 추가한 주문 결과 출력
        print('주문 음료', self.temp, selected_menu[0], ' : ',selected_menu[1], '원')  # 온도 속성을 추가한 주문 결과 출력


        # 첫 주문일 때 처리!
        self.order_menu.append(f'{self.temp} {selected_menu[0]}')   # 가격 리스트에 추가 + 온도도 함께 추가
        self.order_price.append(selected_menu[1])  # 주문 리스트에 추가
        self.price_sum += selected_menu[1]

        # 추가 주문 또는 지불
        while n != 0:  # 지불을 선택하기 전까지 반복합니다.
            print()  # 줄 바꾸면서
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))  # 추가 주문 또는 지불
            if n > 0 and n < len(self.menu) + 1:
                # 추가 음료의 온도를 입력받고, 가격 리스트, 주문 리스트, 합계 금액을 업데이트
                selected_menu = self.menu[n-1]

                while True:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = 'Hot'
                        break
                    elif t == 2:
                        self.temp = 'ICE'
                        break
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")

                #추가 주문건 처리!
                #self.order_menu.append(self.menu[n-1])   # 가격 리스트에 추가
                self.order_menu.append(f'{self.temp} {selected_menu[0]}')   # 가격 리스트에 추가 + 온도도 함께 추가
                self.order_price.append(selected_menu[1])  # 주문 리스트에 추가
                self.price_sum += selected_menu[1]         # 합계 금액 업데이트

                print('추가 주문 음료', self.temp, selected_menu[0], ':', selected_menu[1], '원\n', '합계 : ', self.price_sum, '원')
            else :
                if n == 0 :  # 지불을 입력하면
                    self.order_time = datetime.now()          # 주문시각 (확장 3. datetime으로 주문 일시 추가)
                    print("주문이 완료되었습니다.")
                    print(self.order_menu, self.order_price)  # 확인을 위한 리스트를 출력
                else :  # 없는 번호를 입력할 때
                    print("없는 메뉴입니다. 다시 주문해 주세요.")

    # 지불
    def pay(self):
        while True:
            b = int(input("지불수단을 고르세요 (현금:1, 카드:2): "))
            if b == 1:
                print("직원을 호출하겠습니다.")
                break
            elif b == 2:
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                break
            else:
                print("1과 2 중 하나를 입력하세요.\n")

    # 주문서 외곽 출력 (확장 4. 데코레이터 사용)
    def border(func):
        def wrapper(*args, **kwargs):
            print('⟝' + '-' * 30 + '⟞')
            for i in range(5):
                print('|' + ' ' * 31 + '|')
            result = func(*args, **kwargs)
            for i in range(5):
                print('|' + ' ' * 31 + '|')
            print('⟝' + '-' * 30 + '⟞')
            return result
        return wrapper

    # 주문서 출력 (확장 4. 데코레이터 사용)
    @border
    def table(self):
        # 주문 시각
        print('주문시각 : ', self.order_time.strftime('%Y-%m-%d: %H:%M:%S'))   # datetime의 strftime(format) 메서드
        # 주문 상품명 : 해당 금액
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i],'원')

        print('합계 금액 :', self.price_sum,'원')


### 출력 ###
a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
